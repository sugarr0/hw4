```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('NY.csv')
```

```python
train
```

Кореляция между ценой, площадью помещения, количеством спален и количеством ванных комнат

```python
t = train.copy()
```

```python
t = t.drop(["BROKERTITLE", "TYPE", "ADDRESS", "STATE", "MAIN_ADDRESS", "ADMINISTRATIVE_AREA_LEVEL_2", "LOCALITY", "SUBLOCALITY", "STREET_NAME", "LONG_NAME", "FORMATTED_ADDRESS", "LATITUDE", "LONGITUDE"], axis=1)
```

```python
f,ax = plt.subplots(figsize=(20, 20))
sns.heatmap(t.corr(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)
plt.show()
```

Средняя цена и средняя площадь для разного количества спален

```python

t.groupby('BEDS').mean()
```

Количество квартир по количеству спален:

```python

plt.figure(figsize=(15,10))
sns.countplot(t['BEDS'].sort_values().astype('str'))
plt.title("Количество квартир",fontsize=15)
plt.show()
```
